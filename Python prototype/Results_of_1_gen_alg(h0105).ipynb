```python
import numpy as np
from PIL import Image, ImageOps
import math
import random
import sys, getopt
from scipy.fft import fft2, fftfreq, fftshift, ifft2
from scipy import fftpack, ndimage
from matplotlib import pyplot as plt
import cmath
from scipy.signal import convolve2d
from matplotlib.pyplot import imread
import os
import time
```

```python
def M(x, y):
    if(x**2 + y**2 < (math.pi/2)**2):
        return 1
    else:
        return 0
```

```python
def p_s(x, y, delta_z):
    return 0.375 * abs(delta_z) * math.pi * (x**2 + y**2)
```

```python
def p(x, y):
    return 0.5 * math.pi * (x**2 + y**2)
```

```python
def h_orig(delta_z, sh_orig):  # создает матрицу для фильтра нужной нам размерности (h)
                               # от сигма зависит степень размытия. чем больше сигма тем больше размытие
    
    matr = np.zeros(shape = (sh_orig, sh_orig), dtype = np.complex128) # матрица из 0 (по размеру изборжания подаваемого)
                                          
    #|F(M(x,y)∙exp{i∙p_s (x,y,∆z_nm )+i∙p(x,y)})|
     
    for i in range(matr.shape[0]):
        for j in range(matr.shape[1]):
            x = math.pi/sh_orig * (i - sh_orig/2)
            y = math.pi/sh_orig * (j - sh_orig/2)
            matr[i, j] = M(x,y) * cmath.exp(1j * p_s(x, y, delta_z) + 1j * p(x, y))
    return abs(fftshift(fft2(matr)))**2
```

```python
def read_fft_pics_rash(number_of_pics, sh_orig, part_of_path):
    pics = [0] * number_of_pics
#     path = (os.path.abspath(os.curdir))+'/'+str(sh_orig)+'x'+str(sh_orig)+'/'
    path = os.path.abspath(os.curdir) + part_of_path
    dir_pics = os.listdir(path = path)
    dir_pics.sort()
    
    for i in range(number_of_pics):
        im = Image.open(path + dir_pics[i],'r')

        data = np.array(im)                                                #представляем как массив
        data_im_fl = data[:,:].astype(np.float)
        
        im_f_n = np.zeros(shape = (sh_orig*2,sh_orig*2), dtype = float)
        im_f_n[0:int(sh_orig), 0:int(sh_orig)] = data_im_fl                #расширяем
        
        pics[i] += fft2(im_f_n)
    return pics
```

```python
def h_rash(amount_of_pics, sh_orig, part__of_delta_z):
    mas_of_h = [0] * amount_of_pics
    
    for i in range (0, amount_of_pics):
        h_f_n = np.zeros(shape = (sh_orig*2,sh_orig*2), dtype = np.complex128)
        h_f_n[0:int(sh_orig), 0:int(sh_orig)] = h_orig(delta_z = part__of_delta_z * i, sh_orig = sh_orig)
        h_sum = np.sum(h_f_n)
        for k in range(h_f_n.shape[0]): #нормировка
            for f in range(h_f_n.shape[1]):
                h_f_n[k][f] /= h_sum
        mas_of_h[i] += (fft2(h_f_n))
    return mas_of_h
```

```python
def recovery_func_check(recovered_data, original_data):
    
    original_data_dispercion = original_data / np.var(original_data)
    original_data_average_tmp = original_data_dispercion / np.mean(original_data_dispercion)
    original_data_average = np.zeros(shape = (sh_orig*2, sh_orig*2), dtype = np.float)
    original_data_average[:256,:256] = original_data_average_tmp
#     print(np.shape(original_data_average))
    
    recovered_data_dispercion = recovered_data / np.var(recovered_data)
    recovered_data_average = recovered_data_dispercion / np.mean(recovered_data_dispercion) 
    
    recovered_data_average = fft2(recovered_data_average) / np.max(recovered_data_average)
    original_data_average = fft2(original_data_average) / np.max(original_data_average)
    
    tmp_sum = abs(recovered_data_average / original_data_average)
    res = np.zeros((4, sh_orig))
    res[0] = np.arange(0, sh_orig, 1)
    for r in range(sh_orig-1):
        for u in range (sh_orig*2):
            for v in range (sh_orig*2):
    #             R0 = np.sqrt(u**2 + v**2)      
                if ( (r**2 <=  (u**2 + v**2)) and ((r+1)**2 >= (u**2 + v**2)) ):
                    res[1][r] += tmp_sum[u][v]
                    res[2][r] += 1
                    res[1][r+1] += tmp_sum[u][v]
                    res[2][r+1] += 1
        for i in range(sh_orig):
            if(res[2][i] > 0):
                res[3][i] = res[1][i] / res[2][i]
            else:
                res[3][i] = 0

        rad = res[0]
        A = res[3]
    return rad, A
```
